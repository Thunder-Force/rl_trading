### 0. Import dependencies

```python
# Gym stuff
import gym
import gym_anytrading

# Stable baselines - rl stuff
from stable_baselines.common.vec_env import DummyVecEnv
from stable_baselines import A2C


# Processing libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

### 1. Bring in Marketwatch GME Data 
https://www.marketwatch.com/investing/stock/gme/download-data?startDate=11/1/2019&endDate=03/12/2021

```python
df = pd.read_csv('data/gme.csv')
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.dtypes
```

```python
df.set_index('Date', inplace=True)
df.head()
```

```python
env = gym.make('stocks-v0', df=df, frame_bound=(5,100), window_size=5)
```

```python
env.signal_features
```

### 2. Build Environment

```python
env.action_space
```

```python
state = env.reset()
while True: 
    action = env.action_space.sample()
    n_state, reward, done, info = env.step(action)
    if done: 
        print("info", info)
        break
        
plt.figure(figsize=(15,6))
plt.cla()
env.render_all()
plt.show()
```

### 3. Build Environment and Train

```python
env_maker = lambda: gym.make('stocks-v0', df=df, frame_bound=(5,100), window_size=5)
env = DummyVecEnv([env_maker])
```

```python
model = A2C('MlpLstmPolicy', env, verbose=1) 
model.learn(total_timesteps=1000000)
```

### 4. Evaluation

```python
env = gym.make('stocks-v0', df=df, frame_bound=(10,140), window_size=5)
obs = env.reset()
while True: 
    obs = obs[np.newaxis, ...]
    action, _states = model.predict(obs)
    obs, rewards, done, info = env.step(action)
    #print(obs)
    #print(rewards)
    if done:
        print("info", info)
        break
```
